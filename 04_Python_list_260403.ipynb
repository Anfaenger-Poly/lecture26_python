```python
a_list = ['one', 'two', 'three', 'for']
print('원본 데이터', a_list)

a_list.reverse()
print('역순 후 데이터', a_list)
print('==============================')

a_list.sort()
print('정렬 후 데이터', a_list)
print('==============================')
a_list = ['one', 'two', 'three', 'for']
b_list = sorted(a_list)
print('원본 데이터', a_list)
print('정렬 후 데이터', b_list)
print('==============================')

c_list = sorted(a_list, key=len)
print(c_list)
```

```python
scores = [95, 45, 21, 74, 55]
print(sorted(scores))
print('==============================')
scores = [95, 45, 21, 74, 55]
print(sorted(scores, reverse=True))
```

```python
# 1~N의 수로 이루어진 리스트 생성
N = int(input('숫자 :'))
num_list = []
for i in range(1, N+1):
    num_list.append(i)
print(num_list)
```

```python
# 1~N의 수로 이루어진 리스트 생성
N = int(input('숫자 :'))
num_list = [i for i in range(1, N+1)]
print(num_list)
print('==============================')

# expression
num_list = [i * 2 for i in range(1, N+1)]
print(num_list)
print('==============================')

# if문 사용
N = int(input('숫자 :'))
num_list = [i for i in range(1, N+1) if i % 2 ==0]
print(num_list)
```

### 리스트와 반복구조 연습

```python
n_list = [23, 45, 27, 11, 25, 65, 78]
def getIndex(n_list, target):
    for a in range(len(n_list)):
        if n_list[a] == target:
            return a
    return -1

print(getIndex(n_list, 25))
```

```python
n_list = [23, 45, 27, 11, 25, 65, 78]
def getMax(n_list):
    max_val = n_list[0]
    for a in n_list:
        if a > max_val:
            max_val = a
    return max_val

print(getMax(n_list))
```

```python
n_list = [23, 45, 27, 11, 25, 65, 78]
def getMin(n_list):
    min_val = n_list[0]
    for a in n_list:
        if a < min_val:
            min_val = a
    return min_val

print(getMin(n_list))
```

```python
n_list = [23, 45, 27, 11, 25, 65, 78]
def countGT(n_list):
    count = 0
    target = 42
    for a in n_list:
        if a > target:
            count += 1
    return count

print(countGT(n_list))
```

```python
n_list = [23, 45, 27, 11, 25, 65, 78]
def sumList(n_list):
    total = 0
    for a in n_list:
        total += a
    return total

print(sumList(n_list))
```

```python
n_list = [23, 45, 27, 11, 25, 65, 78]
def swapList(n_list):
    left = 0
    right = len(n_list) - 1

    while left < right:
        n_list[left], n_list[right] = n_list[right], n_list[left]
        left += 1
        right -= 1
        
swapList(n_list)
print(n_list)
```
