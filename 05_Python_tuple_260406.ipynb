### 튜플 (tuple)

```python
t4 = ()
t5 = (1,)
t6 = 1, 2, 3
t7 = 1,

print(t6)
x, y, z = t6
print(x, y, z)
```

```python
# 인덱싱
t = (1, 2, 3, 4)

print(t[0])
print(t[2])
print(t[-1])
```

```python
# 슬라이싱(Slicing)
t = (1, 2, 3, 4, 5)

print(t[1:4])
print(t[:3])
print(t[2:])
print(t[::-1])
```

```python
# for ... in
t = (1, 2, 3)
for a in t:
    print(a)
```

```python
# in 연산자
t = (1, 2, 3)
print(2 in t)
print(5 in t)
```

```python
# index()
t = (1, 2, 3, 2)
print(t.index(2))
```

```python
# count()
t = (1, 2, 3, 2)
print(t.count(2))
print(t.count(4))
```

```python
# len()
t = (1, 2, 3)
print(len(t))
```

```python
# max(), min()
t = (1, 5, 10)
print(max(t))
print(min(t))
```

```python
# sum()
t = (1, 5, 10)
print(sum(t))
```

```python
# sorted()
t = (5, 1, 10)
result = sorted(t)
print(result)
print(tuple(result))
```

```python
# +연산
t1 = (1, 2)
t2 = (3, 4)
t3 = t1 + t2
print(t3)
```

```python
# *연산
t = (1, 2)
t1 = t * 3
print(t1)
```

```python
# 튜플 전체 삭제
t = (1, 2, 3)
del t
print(t)
```

```python
# 항목 추가 append()
lst = [1, 2]
lst.append(3)
print(lst)
```

```python
# insurt()
lst = [1, 2]
lst.insert(1, 50)
print(lst)
```

```python
# extend()
lst = [1, 2]
lst.extend([3, 4])
print(lst)
```

```python
# pop()
lst = [1, 2, 3]
a = lst.pop()
print(a)
print(lst)
```

```python
# removed()
lst = [1, 2, 3, 2]
lst.remove(2)
print(lst)
```

```python
# clear()
lst = [1, 2, 3]
lst.clear()
print(lst)
```

```python
# del 항목
lst = [1, 2, 3]
del lst[1]
print(lst)
```

```python
# reverse()
lst = [1, 2, 3]
lst.reverse()
print(lst)
```

```python
# sort()
lst = [3, 1, 2]
lst.sort()
print(lst)
```

### 튜플의 활용

```python
# 두 변수의 값 바꾸기 (swap)
x, y = 10, 20
x, y = y, x
print(x, y)
```

```python
# 함수에서 여러 개의 값 반환
def add_and_mul(a, b):
    return a+b, a*b
result = add_and_mul(3, 5)
print(result)
```
