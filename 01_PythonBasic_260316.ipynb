### 02. 대입 연산자
- 두 변수의 값을 교환하느 프로그램 (swap)

```python
a = 10
b = 20
print(a, b)
temp = a
a = b
b = temp
print(a,b)
```

```python
a = b = c = 7
print(a, b, c)
```

```python
a, b, c = 7, 11, 19
print(a, b, c)
```

```python
a = 15
b = 25
print(a, b)
a, b = b, a
print(a, b)
```

```python
a = 10 > 20
print(a)
```

```python
a = 10 < 20
print(a)
```
