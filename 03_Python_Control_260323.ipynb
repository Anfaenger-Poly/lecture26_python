```python
# while문 연습 : 나무 찍기

tree_hit = 0
while tree_hit < 10:
    tree_hit += 1
    print(f'나무를 {tree_hit}번 찍었습니다.')
print("나무가 넘어갑니다.")
```

```python
# while문 연습 : 입력값에 따라 반복 종료하기
# 사용자로부터 숫자를 하나씩 입력 받아서 총합과 평균을 출력
# 사용자가 0을 입력하면 숫자 입력을 멈춤

sum = count = 0
num = int(input("숫자 입력 (0 : 끝) :"))
while num != 0:
    sum += num
    count += 1
    num = int(input("숫자 입력 (0 : 끝) :"))
                    

# 0 으로 나누기를 하면 ZeroDivisionError(zero_division_error 발생) -> 예외처리
if count != 0:
    average = sum / count
else:
    average = 0
print(f">> 총합 : {sum}, 평균 : {average:.2f}")
```

```python
# while문 연습 : 입력값에 따라 반복 종료하기
# 사용자로부터 숫자를 하나씩 입력 받아서 총합과 평균을 출력
# 사용자가 0을 입력하면 숫자 입력을 멈춤

sum = count = 0
while True:
    num = int(input("숫자 입력 (0 : 끝) :"))
    if num == 0:
        break
    sum += num
    count += 1

# 0 으로 나누기를 하면 ZeroDivisionError(zero_division_error 발생) -> 예외처리
if count != 0:
    average = sum / count
else:
    average = 0

print(f">> 총합 : {sum}, 평균 : {average:.2f}")
```

```python
# 1부터 N까지의 합 구하기

num = int(input('숫자를 입력하세요 :'))

sum = 0
for num in range(1, num + 1):
# print(num)
    sum += num

print(f'1부터 {num}까지의 합은 {sum}입니다.')
```

```python
# 입력 받은 숫자에 해당하는 구구단 출력

num = int(input('숫자를 입력하세요 :'))

for a in range(1, 10):
       print(num, '*', a, '=', (num * a))

print(f'구구단 {num}단')
```

```python
# 입력받은 숫자의 팩토리얼(Factorial) 값을 출력

num = int(input('숫자를 입력하세요 :'))
result = 1
for num in range(1, num + 1):
    result *= num

print(f'{num}! = {result}')
```
