### BOJ 11478 서로 다른 부분 문자열의 개수

```python
S = input().strip()
result = set()

for i in range(len(S)): # 부분 문자열의 시작 index
    for j in range(i, len(S)): # 부분 문자열의 끝 index
        sub_str = S[i:j+1]
        result.add(sub_str)

print(len(result))
```

```python
S = input().strip()
part = set()

start = 0
end = 1
l = len(S)

while True:
    part.add(S[start:end])
    end += 1
    if end > l:
        start += 1
        end = start + 1
    if start > l:
        break

print(len(part) - 1)
```

```python
import sys
input = sys.stdin.readline

n = int(input())
in_person = set()

for _ in range(n):
    name, action = input().strip().split()
    
    if action == "enter":
        in_person.add(name)
    else:
        in_person.discard(name)

for person in sorted(in_person, reverse=True):
    print(person)
```
